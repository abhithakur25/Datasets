```python
age_med=Titanic_Data['Age'].median()
age_med
```

```python
age_mean=Titanic_Data['Age'].mean()
age_mean
```

```python
# numerical python - solution for mathematical operations
import numpy as np
# replace : 'np.nan' - value to be replaced, age_med = replaced value
Titanic_Data['Age'].replace(np.nan,age_med, inplace=True)
```

```python
Titanic_Data.isnull().sum()
```

```python
Titanic_Data['Embarked'].replace(np.nan,age_med, inplace=True)
Titanic_Data.isnull().sum()
```

```python
# It drops entire row, if any missing values are present
Titanic_Data.dropna(inplace=True)
Titanic_Data.isnull().sum()
```
