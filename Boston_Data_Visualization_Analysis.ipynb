```python
# matplotlib - visualization for simple graph - bar, line, scatter, pie, histogram
import matplotlib.pyplot as plt

# Complex graphs & data distribution - violin plot, pairplot, heatmap etc.
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
boston_data = pd.read_csv('boston_train.csv')
boston_data.head()
```

# Line Graph

```python
# Relation graph b/w x-axis and y-axis
plt.figure(figsize=(10,5))
plt.plot(boston_data['AGE'],color='red')
plt.title('Line Graph')
plt.xlabel('Row Indexes')
plt.ylabel('AGE')
plt.show()
```

```python
boston_data['AGE'].sort_values().unique()
```

```python
# Relation graph b/w x-axis and y-axis
plt.figure(figsize=(10,5))

plt.plot(boston_data['AGE'].sort_values().unique())
plt.title('Line Graph')
plt.xlabel('Row Indexes')
plt.ylabel('AGE')
plt.show()
```

# Histogram

* Distribution of continious Samples

```python
# 0-100, 100/20=5
# (1-5),(6-10),(11-15),(16-20)
plt.hist(boston_data['MEDV'],bins = 20)
plt.title('Histogram (Distribution of Age)', fontsize=9)
plt.xlabel('AGE', fontsize=9)
plt.ylabel('Frequency', fontsize=9)
plt.show()
```

```python
fig = plt.figure(figsize = (10,8))
ax = plt.axes(projection = "3d")

def z_function(x,y):
  return np.cos(np.sqrt(x ** 2 + y ** 2))

## Generate random x & y b/w range -6 & 6 with 30 samples 
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X,Y = np.meshgrid(x,y)
#print(X)
#print(Y)
Z = z_function(X,Y)

ax.plot_surface(X, Y, Z, cmap = 'Reds', edgecolor = 'none')
```
